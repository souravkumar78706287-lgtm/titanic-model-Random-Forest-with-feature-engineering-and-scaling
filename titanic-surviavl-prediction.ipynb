```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

train_df=pd.read_csv("/kaggle/input/competitions/titanic/train.csv")

test_df=pd.read_csv("/kaggle/input/competitions/titanic/test.csv")

test_passenger_id=test_df['PassengerId']

df=pd.concat([train_df,test_df],axis=0)

df['FamilySize']=df['SibSp']+df['Parch']+1

df['IsAlone']=np.where(df['FamilySize']>1,0,1)

df['Title']=df['Name'].str.extract(' ([A-Za-z]+)\.',expand=False)

df['Age']=df['Age'].fillna(df['Age'].median())

df['Fare']=df['Fare'].fillna(df['Fare'].median())

df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])

df['Cabin']=df['Cabin'].fillna('Missing')

categorical_features=['Sex','Embarked','Title','Cabin']

le=LabelEncoder()

for feature in categorical_features:
    df[feature]=le.fit_transform(df[feature])

df.drop(['Name','Ticket'],axis=1,inplace=True)

scaler=StandardScaler()

numerical_features=['PassengerId','Pclass','Age','SibSp','Parch','Fare','FamilySize']

df[numerical_features]=scaler.fit_transform(df[numerical_features])

train_processed=df[df['Survived'].notnull()]

test_processed=df[df['Survived'].isnull()]

X_train=train_processed.drop('Survived',axis=1)

y_train=train_processed['Survived']

X_test=test_processed.drop('Survived',axis=1)

model=RandomForestClassifier(n_estimators=100,random_state=42)

model.fit(X_train,y_train)

predictions=model.predict(X_test)

submission=pd.DataFrame({
    'PassengerId':test_passenger_id,
    'Survived':predictions.astype(int)
})

submission.to_csv('submission.csv',index=False)

print(submission.head())
```
